```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
#Dataset 1
```

```python
#datalgth=6560 max
datalgth=int(input("Enter size of data"))
```

```python
X=np.ndarray((datalgth,3))
Y=np.ndarray((datalgth,1))
lc=0
with open("/home/devansh/Downloads/Datasets-Question1/Train1.csv") as csvfile:
    csv_reader=csv.reader(csvfile, delimiter=',')
    for row in csv_reader:
        X[lc][0]=row[0]
        X[lc][1]=row[1]
        X[lc][2]=1
        Y[lc][0]=row[2]
        if(Y[lc][0]==0):
            Y[lc][0]=-1
        lc=lc+1
```

```python
loss=[]
```

```python
def perceptron(alpha,it):
    We=np.random.random((3,1))
    epoch=0
    for i in range(it):
        lo=0
        for j in range(datalgth):
            if((We[0][0]*X[j][0]+We[1][0]*X[j][1]+We[2][0])*(Y[j][0])<=0):
                lo+=1
                We[0][0]=We[0][0]+((X[j][0]*Y[j][0]))
                We[1][0]=We[1][0]+((X[j][1]*Y[j][0]))
                We[2][0]=We[2][0]+((X[j][2]*Y[j][0]))
        loss.append(lo)
    t=np.linspace(2,18,1000)
    plt.plot(t,(-1)*(We[0][0]*t+We[2][0])/We[1][0],'g-')
    for i in range(datalgth):
        if(Y[i][0]==1):
            plt.plot(X[i][0],X[i][1], 'bo')
        else :
            plt.plot(X[i][0],X[i][1], 'ro')
    plt.show()
    return We
```

```python
W0=perceptron(0.05,150)
```
