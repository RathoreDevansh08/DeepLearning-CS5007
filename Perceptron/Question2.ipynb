```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
datatr=1100
datat=272
```

```python
X=np.ndarray((datatr,5))
Y=np.ndarray((datatr,1))
Xt=np.ndarray((datat,5))
Yt=np.ndarray((datat,1))
lc=0
lc2=0
with open("/home/devansh/Downloads/datasetq2.csv") as csvfile:
    csv_reader=csv.reader(csvfile, delimiter=',')
    for row in csv_reader:
        if(lc<1100):
            X[lc][0]=row[0]
            X[lc][1]=row[1]
            X[lc][2]=row[2]
            X[lc][3]=row[3]
            X[lc][4]=1
            Y[lc][0]=row[4]
            if(Y[lc][0]==0):
                Y[lc][0]=-1
        else:
            Xt[lc2][0]=row[0]
            Xt[lc2][1]=row[1]
            Xt[lc2][2]=row[2]
            Xt[lc2][3]=row[3]
            Xt[lc2][4]=1
            Yt[lc2][0]=row[4]
            if(Yt[lc2][0]==0):
                Yt[lc2][0]=-1
            lc2+=1
        lc=lc+1
```

```python
loss=[]

def perceptron(alpha,it):
    We=np.random.random((5,1))
    epoch=0
    for i in range(it):
        lo=0
        for j in range(datatr):
            if((We[0][0]*X[j][0]+We[1][0]*X[j][1]+We[2][0]*X[j][2]+We[3][0]*X[j][3]+We[4][0])*(Y[j][0])<=0):
                lo+=1
                We[0][0]=We[0][0]+((X[j][0]*Y[j][0]))
                We[1][0]=We[1][0]+((X[j][1]*Y[j][0]))
                We[2][0]=We[2][0]+((X[j][2]*Y[j][0]))
                We[3][0]=We[3][0]+((X[j][3]*Y[j][0]))
                We[4][0]=We[4][0]+((X[j][4]*Y[j][0]))
        loss.append(lo)
    return We
```

```python
W0=perceptron(0.05,500)
```

```python
p=[]
t=[]
for i in range(500):
    p.append(loss[i])
    t.append(i)
plt.plot(t,p)
plt.show()
```

```python
p=[]
t=[]
for i in range(500):
    p.append((datatr-loss[i])/datatr*100)
    t.append(i)
plt.plot(t,p)
plt.show()
```
